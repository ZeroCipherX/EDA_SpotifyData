# Spotify Dataset EDA with Insights



```python
# Import necessary libraries for data manipulation and visualization.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Spotify dataset and view its shape and first few rows.
df = pd.read_csv('spotify_dataset.csv')
print('Shape:', df.shape)
df.head()
```

## Drop columns not needed for EDA

```python
# Drop columns that are IDs or not useful for exploratory analysis.
drop_cols = ['Index', 'Song ID', 'Chord', 'Week of Highest Charting']
df = df.drop(columns=[c for c in drop_cols if c in df.columns])
df.head()
```

## Convert numeric-like columns stored as text

```python
# Convert columns stored as text (with commas, etc.) to numeric type for analysis.
numeric_like = [
    'Streams','Artist Followers','Popularity','Danceability','Energy',
    'Loudness','Speechiness','Acousticness','Liveness','Tempo','Duration (ms)','Valence',
    'Number of Times Charted'
]
for col in numeric_like:
    if col in df.columns:
        s = df[col].astype(str).str.replace(',', '', case=False)
        s = s.str.strip()
        df[col] = pd.to_numeric(s, errors='coerce')
```

```python
# Fix 'Weeks Charted' column: if it's a string with date ranges separated by \n, count the number of weeks for each song.
if 'Weeks Charted' in df.columns:
    df['Weeks Charted'] = df['Weeks Charted'].apply(
        lambda x: len(str(x).split('\n')) if pd.notnull(x) and isinstance(x, str) else pd.to_numeric(x, errors='coerce')
    )
```

## Feature Engineering

```python
# Feature engineering: extract year, create total chart presence, convert duration to minutes.
if 'Release Date' in df.columns:
    df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
    df['Release Year'] = df['Release Date'].dt.year

if 'Weeks Charted' in df.columns and 'Number of Times Charted' in df.columns:
    df['Total Chart Presence'] = df['Weeks Charted'].fillna(0) + df['Number of Times Charted'].fillna(0)

if 'Duration (ms)' in df.columns:
    df['Duration (min)'] = df['Duration (ms)'] / (1000 * 60)
```

```python
# Show summary statistics and check for missing values.
df.isnull().sum()
```

```python
df.describe(include='all')
```

## Handling Missing Values

```python
before_na = df.isna().sum()

df.fillna(df.mean(numeric_only=True), inplace=True)
for col in df.select_dtypes(include=['object', 'category']).columns:
    if df[col].isna().any():
        df[col].fillna(df[col].mode()[0], inplace=True)

# Fill Release Date with median
if 'Release Date' in df.columns:
    date_non_na = df['Release Date'].dropna().sort_values().reset_index(drop=True)
    if len(date_non_na) > 0:
        median_idx = len(date_non_na) // 2
        median_date = date_non_na.iloc[median_idx]
        df['Release Date'].fillna(median_date, inplace=True)

after_na = df.isna().sum()
print('Missing values before:\n', before_na)
print('\nMissing values after:\n', after_na)
```

## Create main EDA DataFrame, **including all audio features**

```python
base_cols = [
    'Song Name', 'Artist', 'Genre', 'Streams', 'Popularity',
    'Release Year', 'Total Chart Presence', 'Duration (min)'
]
audio_features = [
    'Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness',
    'Liveness', 'Tempo', 'Valence'
]
all_cols = base_cols + [col for col in audio_features if col in df.columns]
df_clean = df[all_cols].copy()

# Optional: Mood Score
if all(col in df.columns for col in ['Valence', 'Acousticness']):
    df_clean['Mood Score'] = (df['Valence'] + (1 - df['Acousticness'])) / 2

df_clean.head()
```

## Replace zero values where zero is likely invalid

```python
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
zero_counts = (df_clean[numeric_cols] == 0).sum()
print("Zero values per numeric column:\n", zero_counts)

# Replace zero with median for columns where zero is likely invalid
for col in numeric_cols:
    if col not in ['Streams', 'Popularity']:
        df_clean[col] = df_clean[col].replace(0, df_clean[col].median())
```

## Visualizations
### Use df_clean for all plots

```python
# 1. Distribution of Streams
plt.figure(figsize=(8,5))
sns.histplot(df_clean['Streams'], bins=50, kde=True)
plt.xscale("log")
plt.title("Distribution of Song Streams (log scale)")
plt.xlabel("Streams (log scale)")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Popularity vs Streams
plt.figure(figsize=(7,5))
sns.scatterplot(data=df_clean, x='Streams', y='Popularity', alpha=0.5)
plt.xscale("log")
plt.title("Popularity vs Streams")
plt.show()
```

```python
# 3. Correlation Heatmap of Audio Features (now using df_clean)
plt.figure(figsize=(10,6))
audio_for_corr = [col for col in audio_features + ['Popularity'] if col in df_clean.columns]
corr = df_clean[audio_for_corr].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Between Popularity and Audio Features")
plt.show()
```

```python
# 4. Top 10 Artists by Total Streams
if 'Artist' in df_clean.columns and 'Streams' in df_clean.columns:
    top_artists = df_clean.groupby('Artist')['Streams'].sum().nlargest(10)
    plt.figure(figsize=(8,5))
    sns.barplot(x=top_artists.values, y=top_artists.index, palette="magma")
    plt.title("Top 10 Artists by Total Streams")
    plt.xlabel("Total Streams")
    plt.show()
```

```python
# 5. Average Duration of Songs
plt.figure(figsize=(7,5))
sns.histplot(df_clean['Duration (min)'], bins=30, kde=True, color="teal")
plt.axvline(df_clean['Duration (min)'].mean(), color='red', linestyle='--', label="Mean Duration")
plt.title("Distribution of Song Duration (in minutes)")
plt.xlabel("Duration (min)")
plt.legend()
plt.show()
```

## Spotify EDA Summary

This analysis started by cleaning up Spotify track data and creating a few new features like mood score and chart presence to add more depth. I then explored the numbers and visualized trends, looking at streams, popularity, and different musical attributes.

A couple of interesting things stood out: most songs have modest stream counts with just a handful getting massive numbers. While popularity and stream counts usually go hand-in-hand, there are definitely some tracks that buck the trend. The audio features of top hits, like danceability and energy, tend to cluster around certain values, hinting at what listeners enjoy most. After the cleaning and prep, the dataset feels solid and trustworthy for more advanced analysis.

All of this matters because it gives artists, record labels, and streaming platforms real insight into what makes a song successful. These findings can shape playlists, marketing strategies, and even future releases. Plus, the data is now in great shape for more complex modeling, like predicting the next big hit or grouping songs by style.

In short, this EDA turns raw Spotify numbers into useful, human-centered insights about what people actually want to hear.
