# Spotify Dataset EDA with Insights

This notebook performs a complete EDA on the Spotify dataset, including:
- **Cleaning & type conversion**
- **Dropping unnecessary columns**
- **Merging/engineering columns**
- **Descriptive stats & visualizations**
- **Key insights** 

```python
# 1) Import libraries and loading dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)

df = pd.read_csv('spotify_dataset.csv')
print('Shape:', df.shape)
df.head()
```

## Why drop columns?

**Drop Some Unnecessary Columns:**
- `Index` → not informative for analysis.
- `Song ID` → not useful for trends.
- `Chord` → out-of-scope for popularity/streams EDA.
- `Week of Highest Charting` → that are messy and not needed for our goals.



```python
# Columns to drop
drop_cols = ['Index', 'Song ID', 'Chord', 'Week of Highest Charting']
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

# Convert numeric-like columns stored as text
numeric_like = [
    'Streams','Artist Followers','Weeks Charted','Popularity','Danceability','Energy',
    'Loudness','Speechiness','Acousticness','Liveness','Tempo','Duration (ms)','Valence'
]
for col in numeric_like:
    if col in df.columns:
        # Literal comma removal, strip spaces; then safe numeric conversion
        s = df[col].astype(str).str.replace(',', '', case=False)
        s = s.str.strip()
        df[col] = pd.to_numeric(s, errors='coerce')
```

**Merge / Feature Engineer:**
- `Release Date` → extract `Release Year` for easier trend plots over time.
- `Weeks Charted` + `Number of Times Charted` → create `Total Chart Presence`(more presence → likely more exposure).
- `Duration (ms)` → also provide `Duration (min)` for human-readable length.
​

```python
# Parse Release Date and engineer Release Year
if 'Release Date' in df.columns:
    df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
    df['Release Year'] = df['Release Date'].dt.year

# Engineer Total Chart Presence = Weeks Charted + Number of Times Charted
if 'Weeks Charted' in df.columns and 'Number of Times Charted' in df.columns:
    df['Total Chart Presence'] = df['Weeks Charted'].fillna(0) + df['Number of Times Charted'].fillna(0)

# Duration in minutes for readability
if 'Duration (ms)' in df.columns:
    df['Duration (min)'] = df['Duration (ms)'] / (1000 * 60)

print('After cleaning/engineering:')
df.info()
```

### Handling Missing Values
- Numeric columns → filled with column mean (simple, keeps distribution centered).
- Categorical columns → filled with mode (most frequent value).

This keeps all rows for EDA while avoiding bias from arbitrary constants.

```python
# 3) Handle missing values
before_na = df.isna().sum()

df.fillna(df.mean(numeric_only=True), inplace=True)
for col in df.select_dtypes(include='object').columns:
    if df[col].isna().any():
        df[col].fillna(df[col].mode()[0], inplace=True)

after_na = df.isna().sum()
print('Missing values before:\n', before_na)
print('\nMissing values after:\n', after_na)
```

```python
# Weeks Charted is fully missing → useless for analysis → drop it
if 'Weeks Charted' in df.columns:
    df.drop(columns=['Weeks Charted'], inplace=True)
```

```python
# filling with a median year for better trend alignment
df['Release Date'].fillna(df['Release Date'].median(), inplace=True)
```

```python
after_na = df.isna().sum()
print('\nMissing values after:\n', after_na)
```

```python
# Create simplified dataset
df_clean = df[['Song Name', 'Artist', 'Genre', 'Streams', 'Popularity',
               'Release Year', 'Total Chart Presence', 'Duration (min)']].copy()

# Optional: create a simple "Mood Score" using Valence and Acousticness
if all(col in df.columns for col in ['Valence', 'Acousticness']):
    df_clean['Mood Score'] = (df['Valence'] + (1 - df['Acousticness'])) / 2

df_clean.head()
```

```python
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
zero_counts = (df_clean[numeric_cols] == 0).sum()
print("Zero values per numeric column:\n", zero_counts)
```

```python
# Replace 0 with median (if zero is invalid)
for col in numeric_cols:
    df_clean[col] = df_clean[col].replace(0, df_clean[col].median())
```
